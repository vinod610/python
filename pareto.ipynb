```python
import numpy as np
import seaborn as sns
import scipy.stats as stats
import matplotlib.pyplot as plt
import pylab
m=1000000
a=1
c=list()
paret = (np.random.pareto(a,50000))+10000
for i in range(1,101):
    b = np.percentile(paret,i)
    c.append(b)
plt.figure()
plt.subplot(2,2,1)
plt.title('percentiles')
stats.probplot(c,dist='pareto',sparams=(10),plot=pylab)
plt.grid()
plt.subplot(2,2,2)
plt.title('normal pareto values')
stats.probplot(paret,dist='pareto',sparams=(10),fit='true',plot=plt)
plt.grid()
#plt.subplot(2,2,3)
#sns.kdeplot(paret,bw=0.5)
#plt.title("pdf")
plt.subplot(2,2,3)
kwargs={'cumulative':True}
sns.distplot(paret,hist_kws=kwargs,kde_kws=kwargs)
plt.title("cdf")
plt.show()
```
