```python
import pylab
import numpy as np
import scipy.stats as stats
std_normal=np.random.normal(loc=0,scale=1,size=1000)
print(std_normal)
for i in range(0,101):
    print(i,np.percentile(std_normal,i))
```

```python
measure=np.random.normal(loc=12,scale=20,size=1000)
stats.probplot(measure,dist='norm',plot=pylab)
pylab.show()
```

```python
mea=np.random.uniform(low=10,high=12,size=800000)
stats.probplot(mea,dist='norm',plot=pylab)
pylab.show()
```
