```python
import scipy.stats as stats
import pylab
import numpy as np
import matplotlib.pyplot as plt
fig=plt.figure()
a2=fig.add_subplot(2,1,1)
a=stats.loggamma.rvs(5,size=500)+5
y=stats.probplot(a,dist=stats.norm,plot=a2)
a2.set_title("without boxcox")
a3=fig.add_subplot(2,1,2)
a4, t=stats.boxcox(a)
a5=stats.probplot(a4,dist=stats.norm,plot=a3)
a3.set_title('with boxcox')
pylab.show()
```
